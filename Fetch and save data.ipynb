```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
filename = 'up2007'
response = requests.get('https://myneta.info/{}/index.php?action=show_winners'.format(filename))
```

```python
page = BeautifulSoup(response.text)
page
```

```python
data_table = page.find_all('table')[2]
data = BeautifulSoup(str(data_table)).find_all('tr')
data
```

```python
relevant_data = data[2:]
```

```python
list(relevant_data[0])
```

```python
clean_data = []
columns = data[1].text.split('\n')[2:-1]
columns = [column.strip() for column in columns]
for entry in relevant_data:
    lines = list(entry)
    name = BeautifulSoup(str(lines[3])).td
    data_point = [name.text, lines[4].text, lines[6].text, lines[8].text, lines[10].text, lines[12].text, lines[14].text]
    clean_data.append(data_point)
```

```python
clean_data
```

```python
columns
```

```python
df = pd.DataFrame(clean_data, columns=columns)
```

```python
df.head()
```

```python
def get_number(string):
    return int(string.split('~')[0][3:-1].replace(',', ''))
```

```python
df['Total Assets'] = df['Total Assets'].apply(get_number)
df['Liabilities'] = df['Liabilities'].apply(get_number)
df['Criminal Case'] = df['Criminal Case'].apply(int)
```

```python
df.head()
```

```python
df.to_csv('data/{}.csv'.format(filename))
```
